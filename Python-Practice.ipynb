```python
# Reversing an array

def reverse_array(arr):
    left = 0
    right = len(arr) - 1

    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1

    return arr

print('Hi')
def main():
    array = [1, 2, 3, 4, 5]
    reversed_array = reverse_array(array)
    print("Reversed Array:", reversed_array)
    
    eee
```
